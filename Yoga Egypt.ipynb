## Importing Libraries

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup 
import csv
import re
```

## Exporting data to a csv file

```python
#Opening excel file
file = open('Assignment_ans.csv','a')

writer=csv.writer(file)

#Taing the Headers
data = np.array(['Title','Brand','Link','Model Number','Current Price','Old Price','Saving Price','Sold By','Free Delivery By','Order In','Rating'])

#Writing the headers
writer.writerows([data])

#Looping for all 64 pages
for i in range(1,64):
    #taking url for pages
    url = 'https://www.noon.com/egypt-en/sports-and-outdoors/exercise-and-fitness/yoga-16328/?limit=50&page='+str(i)+'&sort%5Bby%5D=popularity&sort%5Bdir%5D=desc'    
    page = requests.get(url)

    soup = BeautifulSoup(page.content ,'html.parser')
    lists = soup.find_all('span', class_='sc-5e739f1b-0 gEERDr wrapper productContainer')

    for list in lists:
        count = 1
        #Opening the Item Link
        links = 'https://www.noon.com'+list.find('a')['href']
        page_opened = requests.get(links)
        new_soup = BeautifulSoup(page_opened.content, 'html.parser')
        
        
        #Taking the Title of the Item
        try:
            title = new_soup.find('h1', class_='sc-c44e3e2d-12 kODUYj').text
        except:
            title = ''

        #Taking Rating
        try:
            rating = list.find('span', class_='ratingValue').text
        except:
            rating = ''


        #Taking Brand
        try:
            brand = new_soup.find('div', class_='sc-c44e3e2d-4 kmpRqA').text
        except:
            brand = ''

        #Taking Model Number of Product
        try:
            model_number = new_soup.find('div', class_='modelNumber').text.replace('Model Number : ','')
        except:
            model_number = ''


        #Taking Old Price
        try:
            old_price = new_soup.find('div', class_='priceWas').text
        except:
            old_price = ''


        #Taking Current Price
        try:
            curr_price = new_soup.find('div', class_='priceNow').text
        except:
            curr_price = ''

        #Takin Saving Price
        try:
            sav_price = new_soup.find('div', class_='priceSaving').text
        except:
            sav_price = ''


        #free delivery by
        try:
            free_del = new_soup.find('span',class_='highlight').text
        except:
            free_del = ''

        #sold by
        try:
            sold_by = new_soup.find('div',class_='sc-d711b2ac-2 khyqbb').text
        except:
            sold_by = ''

        #Order In
        try:
            order_in = new_soup.find('div', class_='estimator_left').text.replace('Order in ','')[:13].replace('F','')
        except:
            order_in = ''
        
        
        #Putting all the data to a array
        info = [title, brand, links, model_number, curr_price, old_price, sav_price, sold_by, free_del, order_in, rating]
        
        
        #Writing Data into excel file
        writer.writerows([info])
        print(info)
file.close()
```
